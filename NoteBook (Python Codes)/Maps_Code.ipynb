# Concentration Map

```python
import folium
from folium.plugins import FastMarkerCluster
coords = df[['LATITUDE','LONGITUDE']].dropna().values.tolist()

m = folium.Map(
    location=[37.25, -119.5],     
    zoom_start=6,
    tiles='CartoDB Positron'       
)

FastMarkerCluster(data=coords).add_to(m)

if coords:
    lats = [c[0] for c in coords]
    lons = [c[1] for c in coords]
    m.fit_bounds([[min(lats), min(lons)], [max(lats), max(lons)]])

m.save("california_sites_map.html")
print("Saved: california_sites_map.html")
```

# Heat Map

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

df = pd.read_csv("gwdata.csv", encoding='latin1')

df['LATITUDE'] = pd.to_numeric(df['LATITUDE'], errors='coerce')
df['LONGITUDE'] = pd.to_numeric(df['LONGITUDE'], errors='coerce')

df = df.dropna(subset=['LATITUDE', 'LONGITUDE'])


print(df[['LATITUDE', 'LONGITUDE']].head())

if df.empty:
    raise ValueError("No valid latitude/longitude values found in dataset.")

map_center = [df['LATITUDE'].mean(), df['LONGITUDE'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

heat_data = df[['LATITUDE', 'LONGITUDE']].values.tolist()
HeatMap(heat_data, radius=15).add_to(m)

m.save("contamination_heatmap.html")
print("✅ Heatmap saved — open 'contamination_heatmap.html' in your browser.")
```
