### Filtering date

```python
ground_df['LEAK_REPORTED_DATE'] = pd.to_datetime(ground_df['LEAK_REPORTED_DATE'], errors='coerce')

ground_df = ground_df[
    (ground_df['LEAK_REPORTED_DATE'].dt.year >= 1970) &
    (ground_df['LEAK_REPORTED_DATE'].dt.year <= 2024)
                                                        ]

ground_df = ground_df.sort_values(by='LEAK_REPORTED_DATE')
```

### How long it took to close/repair leaks?

```python
ground_df['BEGIN_DATE'] = pd.to_datetime(ground_df['BEGIN_DATE'], errors='coerce')
ground_df['NO_FURTHER_ACTION_DATE'] = pd.to_datetime(ground_df['NO_FURTHER_ACTION_DATE'], errors='coerce')
ground_df['NO_FURTHER_ACTION_DATE'] = ground_df['NO_FURTHER_ACTION_DATE'].dt.tz_localize(None)
ground_df['RESOLUTION_DAYS'] = (ground_df['NO_FURTHER_ACTION_DATE'] - ground_df['BEGIN_DATE']).dt.days
```

```python
ground_df['RESOLUTION_DAYS'].unique
```

```python
ground_df['RESOLUTION_DAYS'].isnull().sum()
```

```python
ground_df['RESOLUTION_DAYS'].mean()
```

```python
ground_df['RESOLUTION_DAYS'].median()
```

```python
ground_df['RESOLUTION_YEARS'] = ground_df['RESOLUTION_DAYS'] / 365

plt.figure(figsize=(8, 5))
sns.boxplot(x=ground_df['RESOLUTION_YEARS'].dropna(), color='lightblue')

median_value_years = ground_df['RESOLUTION_YEARS'].dropna().median()
plt.axvline(median_value_years, color='red', linestyle='--', linewidth=2, label=f"Median = {median_value_years:.2f} years")

plt.title('Resolution Time for Contamination Cases', fontsize=14)
plt.xlabel('Resolution Time (Years)', fontsize=12)
plt.legend()
plt.savefig('Resolution_Time_For_Contamination_Cases_Years.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
ground_df['RESOLUTION_DAYS'].dropna()

plt.figure(figsize=(8, 5))
sns.boxplot(x=ground_df['RESOLUTION_DAYS'].dropna(), color='lightblue')

median_value = ground_df['RESOLUTION_DAYS'].dropna().median()
plt.axvline(ground_df['RESOLUTION_DAYS'].dropna().median(), color='red', linestyle='--', linewidth=2, label=f"Median = {median_value} days")

plt.title('Resolution Time for Contamination Cases', fontsize=14)
plt.xlabel('Resolution Time (Days)', fontsize=12)
plt.legend()
plt.savefig('Resolution Time For Contamination Cases.png', dpi=300, bbox_inches='tight')
plt.show()
```

### When did the leak actually start?

```python
ground_df['Year'] = ground_df['LEAK_REPORTED_DATE'].dt.year
```

```python
ground_df['Year']
```

```python
ground_df['Year'] = ground_df['LEAK_REPORTED_DATE'].dt.year
yearly_trend = ground_df.groupby('Year').size()
plt.figure(figsize=(12,6))
plt.plot(yearly_trend.index, yearly_trend.values, marker='o', color='teal')
peak_year = yearly_trend.idxmax()
peak_value = yearly_trend.max()
lowest_year = yearly_trend.idxmin()
lowest_value = yearly_trend.min()
plt.scatter(peak_year, peak_value, color='red', zorder=5, label=f'Peak Year: {peak_year} ({peak_value} cases)')
plt.scatter(lowest_year, lowest_value, color='orange', zorder=5, label=f'Lowest Year: {lowest_year} ({lowest_value} cases)')
plt.title('Contamination Cases Reported (1970–2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Cases', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig("Contamination Cases Reported", dpi=300, bbox_inches='tight')
plt.show()
```
